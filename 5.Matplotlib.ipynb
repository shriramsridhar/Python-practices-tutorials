```python
#Line plot
import matplotlib.pyplot as plt
year =[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.plot(year,pop)
plt.show()
```

```python
#Scatter plot
import matplotlib.pyplot as plt
year =[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.scatter(year,pop)
plt.show()
```

```python
GDP=[138910,782,105689,55989,3558]
L_Exp=[76,47,83,81,54]
plt.scatter(GDP,L_Exp)
plt.xscale('log')    #Changes x-scale to logarithmic
plt.show()
```

```python
#Histogram
a=[100,100,45,50,60,60,30,30,30,50,75]
plt.hist(a,bins=5)
plt.show()
```

```python
#Line plot Customization
import matplotlib.pyplot as plt
year =[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
year=[1910,1930]+year
pop=[1.56,2.1]+pop
plt.plot(year,pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population')
plt.yticks([0,2,4,6,8],['0','2 bill','4 bill','6 bill','8 bill'])
plt.show()
```

```python
countries=['Qatar','Congo','Singapore','Sweden','India','China']
#col={'Qatar':'red','Congo','green','Singapore','blue','Sweden':'yellow','India':'pink','China':'black'}
GDP=[138910,782,105689,55989,3558,8000]
L_Exp=[76,47,83,81,54,68]
pop=[13,168,56,78,1200,1250]
plt.scatter(GDP,L_Exp,s=pop, alpha=0.6)
plt.xscale('log')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.title('GDP Vs Life Expectancy Vs Population')
plt.text(3558,54,'India',plt.grid(True))
plt.text(8000,68,'China',plt.grid(True))

plt.show()
```
